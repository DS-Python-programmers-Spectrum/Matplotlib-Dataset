```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Mall_Customers.csv")
```

```python
df
```

```python
plt.figure()
plt.hist(df["Age"], bins=5)
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure()
plt.scatter(df["Age"], df["Spending Score (1-100)"])
plt.xlabel("Age")
plt.ylabel("Spending Score")
plt.title("Age vs Spending Score")
plt.show()
```

```python
plt.figure()
plt.plot(df["Spending Score (1-100)"])
plt.xlabel("Customer Index")
plt.ylabel("Spending Score")
plt.title("Line Plot of Spending Score")
plt.show()
```

```python

# Bar chart
gender_counts = df["Genre"].value_counts()

plt.figure()
plt.bar(gender_counts.index, gender_counts.values)
plt.xlabel("Gender")
plt.ylabel("Number of Customers")
plt.title("Gender Distribution")
plt.show()
```

```python
# Bar Chart: Average Spending by Gender
avg_spending = df.groupby("Genre")["Spending Score (1-100)"].mean()

plt.figure()
plt.bar(avg_spending.index, avg_spending.values)
plt.xlabel("Genre")
plt.ylabel("Average Spending Score")
plt.title("Average Spending Score by Gender")
plt.show()
```

```python
# Pie chart
gender_counts = df["Genre"].value_counts()

plt.figure()
plt.pie(gender_counts.values, labels=gender_counts.index, autopct='%1.1f%%')
plt.title("Gender Distribution")
plt.show()
```

```python
# Count values
counts = df["Genre"].value_counts()

# Count plot
plt.figure()
plt.bar(counts.index, counts.values)
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Count Plot of Gender")
plt.show()
```

```python
age_counts = df["Age"].value_counts().sort_index()

plt.figure()
plt.bar(age_counts.index, age_counts.values)
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Count Plot of Age")
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# 1. Age distribution
axs[0, 0].hist(df["Age"], bins=10)
axs[0, 0].set_title("Age Distribution")

# 2. Annual Income distribution
axs[0, 1].hist(df["Annual Income (k$)"], bins=10)
axs[0, 1].set_title("Annual Income Distribution")

# 3. Spending Score distribution
axs[1, 0].hist(df["Spending Score (1-100)"], bins=10)
axs[1, 0].set_title("Spending Score Distribution")

# 4. Gender count (Count Plot)
gender_counts = df["Genre"].value_counts()
axs[1, 1].bar(gender_counts.index, gender_counts.values)
axs[1, 1].set_title("Gender Count")

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 2)

axs[0, 0].set_title("Age Distribution")
axs[0, 1].set_title("Income Distribution")
axs[1, 0].set_title("Spending Score")
axs[1, 1].set_title("Gender Count")

plt.show()
```
